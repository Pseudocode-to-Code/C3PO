```python
# Strict accuracy
import numpy as np

def exact_match(test_df, truth_col, pred_col):
    strict_bools = test_df.apply(lambda row: np.array_equal(row[truth_col], row[pred_col]), axis=1)
    strict_accuracy = strict_bools.sum() / len(test_df)
    return strict_accuracy
```
